### Customers who paid us the most

```python
%sql SELECT \
        c.customer_id, c.first_name, c.last_name, SUM(i.unit_price) total \
     FROM \
        dwh.invoice_fact i LEFT JOIN dwh.customer_dim c ON i.customer_id = c.customer_id \
     GROUP BY \
        c.customer_id, c.first_name, c.last_name \
     ORDER BY \
        SUM(i.unit_price)  DESC \
     LIMIT 10;
```

### Countries sales

```python
%sql SELECT \
    billing_country AS country, COUNT(*) AS total_sales, SUM(unit_price) AS revenue \
FROM \
    dwh.invoice_fact fact \
    LEFT JOIN dwh.invoice_dim invoice ON fact.invoice_id = invoice.invoice_id \
GROUP BY \
    billing_country, unit_price \
ORDER BY \
    total_sales DESC;
```

```python
%sql SELECT \
billing_country AS country, track.genre, SUM(fact.unit_price) AS revenue \
FROM \
    dwh.invoice_fact fact \
    LEFT JOIN dwh.invoice_dim invoice ON fact.invoice_id = invoice.invoice_id \
    LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
    billing_country, track.genre, fact.unit_price \
ORDER BY \
    country, revenue DESC;
```

### Top Hit songs

```python
%sql SELECT \
    track.name, track.artist, COUNT(*) AS total \
FROM \
    dwh.invoice_fact fact \
    LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
    track.name, track.artist \
HAVING \
    COUNT(*) >=3 \
ORDER BY \
    COUNT(*) DESC;
```

### Who is the most popular artist in january 2012 

```python
%sql SELECT \
    track.artist, COUNT(*) AS total_songs \
FROM \
    dwh.invoice_fact fact \
    LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
    LEFT JOIN dwh.date_dim d ON fact.date_id = d.date_id \
WHERE \
    d.year=2012 AND d.month=1 \
GROUP BY \
    track.artist \
ORDER BY \
    COUNT(*) DESC \
LIMIT 1; 
```

## Rollup Example

```python
%sql SELECT \
	track.genre, \
	track.artist, \
	COUNT(*) AS total \
FROM \
	dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	track.genre, \
	track.artist \
UNION ALL \
SELECT \
	track.genre, \
	NULL, \
	COUNT(*) \
FROM \
	dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	track.genre \
UNION ALL \
SELECT \
	NULL, \
	NULL, \
	COUNT(*) \
FROM \
	dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	() \
ORDER BY 1 NULLS LAST, 2 NULLS LAST;
```

```python
# CTE version, faster than previous one

%sql WITH temp1 AS ( \
	SELECT \
		track.genre, \
		track.artist, \
		COUNT(*) AS total \
	FROM \
		dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	track.genre, \
	track.artist \
), \
temp2 AS ( \
	SELECT \
		track.genre, \
		NULL, \
		COUNT(*) \
	FROM \
		dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	track.genre \
), \
temp3 AS ( \
	SELECT \
		NULL, \
		NULL, \
		COUNT(*) \
	FROM \
		dwh.invoice_fact fact \
	LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	() \
) \
SELECT \
	* \
FROM \
	temp1 \
UNION ALL \
SELECT \
	* \
FROM \
	temp2 \
UNION ALL \
SELECT \
	* \
FROM \
	temp3 \
ORDER BY \
	1 NULLS LAST, \
	2 NULLS LAST;
```

### Artist Chart of all time

```python
%sql SELECT \
	track.artist, \
	COUNT(*) AS total_sales, \
	DENSE_RANK() OVER(ORDER BY COUNT(*) DESC) AS rank \
FROM \
	dwh.invoice_fact fact \
	INNER JOIN dwh.track_dim track ON fact.track_id = track.track_id \
GROUP BY \
	track.artist;
```

### Artists who have "0" sales

```python
%sql WITH ranking AS ( \
	SELECT \
		track.artist, \
		COUNT(*) AS total_sales, \
		DENSE_RANK() OVER (ORDER BY COUNT(*) \
			DESC) AS rank \
	FROM \
		dwh.invoice_fact fact \
		INNER JOIN dwh.track_dim track ON fact.track_id = track.track_id \
	GROUP BY \
		track.artist \
) \
SELECT \
	track.artist, \
	COUNT(*) AS total_songs, \
	COALESCE(rank.rank, NULL) AS ranking \
FROM \
	dwh.track_dim track \
LEFT JOIN ranking rank ON track.artist = rank.artist \
WHERE \
	rank.rank is NULL \
GROUP BY \
	track.artist, \
	rank.rank \
ORDER BY \
	rank.rank NULLS FIRST;
```

```python
# List of artists who couldn't make it :(

%sql SELECT \
	t.artist \
FROM \
	dwh.track_dim t \
WHERE \
	NOT EXISTS ( \
		SELECT \
			track.artist \
		FROM \
			dwh.invoice_fact fact \
		LEFT JOIN dwh.track_dim track ON fact.track_id = track.track_id \
	WHERE track.artist=t.artist \
	GROUP BY \
		track.artist) \
GROUP BY \
	t.artist ;
```

### Convert milliseconds to minutes

```python
%sql DROP FUNCTION IF EXISTS dwh.convert_milliseconds;
```

```python
%sql CREATE OR REPLACE FUNCTION dwh.convert_milliseconds(milliseconds BIGINT)RETURNS FLOAT AS $$ DECLARE seconds FLOAT:=milliseconds::double precision/1000;result FLOAT;BEGIN SELECT FLOOR(seconds/60)+(seconds::INT%60.00)/100 INTO result;RETURN result;END;$$ LANGUAGE plpgsql 
```

```python
%sql SELECT name, dwh.convert_milliseconds(milliseconds) AS duration FROM dwh.track_dim;
```

### The longest song

```python
%sql SELECT \
	name, artist, album, dwh.convert_milliseconds(milliseconds) AS duration3 \
FROM \
	dwh.track_dim \
WHERE \
	milliseconds = ( \
		SELECT \
			MAX(milliseconds) \
		FROM \
			dwh.track_dim);
```

### How many songs have length more than 1.40

```python
%sql SELECT \
	COUNT(CASE WHEN LENGTH(milliseconds::TEXT) >= 6 THEN \
		1 \
	ELSE \
		0 \
	END) AS total \
FROM \
	dwh.track_dim;
```

### Count numbers of minutes 

```python
%sql SELECT \
	FLOOR((milliseconds::double precision / 1000) / 60) AS minute, COUNT(*) \
FROM \
	dwh.track_dim \
GROUP BY  \
	minute \
ORDER BY COUNT(*) DESC;
```
